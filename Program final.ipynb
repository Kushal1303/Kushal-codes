```python
"Program to Predict the best artificial lift that can be applied on the basis of input"
from IPython.display import clear_output

gl=[0,"Gas Lift"]
pcp=[0,"Progressive cavit pumping"]
esp=[0,"Electrical Submerssible Pump"]
srp=[0,"Sucker rod pump"]

def selectdepth(depth):                #Depth Testing Function
    global gl,pcp,esp,srp
    if(depth<15000):
        gl[0]=gl[0]+1
        
    if(15000>depth):
        esp[0]=esp[0]+1
    
    if(6000>depth):
        pcp[0]=pcp[0]+1
        
    if(16000>depth):
        srp[0]=srp[0]+1
        
    return

def selectapi(api):                   #API testing function
    global gl,pcp,esp,srp
    if(15.0<api):
        gl[0]=gl[0]+1
        
    if(10.0<api):
        esp[0]=esp[0]+1
    
    if(api<20):
        pcp[0]=pcp[0]+1
        
    if(50>api):
        srp[0]=srp[0]+1
        
    return

def selectba(ba):                   #Build Angle testing Function
    global gl,pcp,esp,srp
    if(ba<70.0):
        gl[0]=gl[0]+1
        
    if(ba<40.0):
        esp[0]=esp[0]+1
    
    if(ba<15.0):
        pcp[0]=pcp[0]+1
        
    if(15.0>ba):
        srp[0]=srp[0]+1
        
    return

def selecttemp(temp):               #Temperature testing Fumction 
    global gl,pcp,esp,srp
    if(400.0>temp):
        gl[0]=gl[0]+1
        
    if(400.0>temp):
        esp[0]=esp[0]+1
    
    if(250.0>temp):
        pcp[0]=pcp[0]+1
        
    if(200>temp):
        srp[0]=srp[0]+1
        
    return

def selectvol(vol):               #volume testing function
    global gl,pcp,esp,srp
    if(vol<6000):
        gl[0]=gl[0]+1
        
    if(30000>vol):
        esp[0]=esp[0]+1
    
    if(4500>vol):
        pcp[0]=pcp[0]+1
        
    if(5000>vol):
        srp[0]=srp[0]+1
        
    return

def selectgor(gor):               #GOR testing function
    global gl,pcp,esp,srp
    if(gor<3000):
        gl[0]=gl[0]+1
        
    if(2000>gor):
        esp[0]=esp[0]+1
    
    if(gor<500):
        pcp[0]=pcp[0]+1
        
    if(gor<500):
        srp[0]=srp[0]+1
        
    return

def selectwc(wc):               #Water cut testing function
    global gl,pcp,esp,srp
    
    if(wc<100):
        gl[0]=gl[0]+1
        
    if(wc<50):
        esp[0]=esp[0]+1 
    
    if(wc<80):
        pcp[0]=pcp[0]+1
        
    if(wc<100):
        srp[0]=srp[0]+1
        
    return

def selectsolid(solid):               #Solid handling  testing function
    global gl,pcp,esp,srp
    
    if(solid<1500):
        gl[0]=gl[0]+1
        
    if(solid<200):
        esp[0]=esp[0]+1
    
    if(2000>solid):
        pcp[0]=pcp[0]+1
        
    if(400>solid):
        srp[0]=srp[0]+1
        
    return

def selectvisco(visco):                #viscosity Testing Function
    global gl,pcp,esp,srp
    if(visco<800):
        gl[0]=gl[0]+1
        
    if(visco<400):
        esp[0]=esp[0]+1
    
    if(visco>1000):
        pcp[0]=pcp[0]+1
        
    if(visco<400):
        srp[0]=srp[0]+1
        
    return

def selectgas(gas):               #Solid handling  testing function
    global gl,pcp,esp,srp
    
    if(gas<100):
        gl[0]=gl[0]+1
        
    if(gas<10):
        esp[0]=esp[0]+1
    
    if(gas==0):
        pcp[0]=pcp[0]+1
        
    if(gas<20):
        srp[0]=srp[0]+1
        
    return

depth=int(input ( "Enter the Depth in ft:"))    
selectdepth( depth )

api=float(input ( "Enter the API: "))    
selectapi( api )

ba=float(input ( "Enter the build angle : "))
selectba( ba )

temp=float(input ( "Enter the down hole temperature in fahrenheit: "))
selecttemp( temp ) 

vol=int(input( "Enter the opearting volume in BOPD:"))
selectvol( vol )

gor=int(input( "Enter the GOR: "))
selectgor( gor )

wc=int(input( "Enter the Water cut %: "))
selectwc( wc )

solid=int(input( "Enter the Solid content in ppm: "))
selectvol( solid )

visco=int(input( "Enter the fuild viscosity in cp:  "))
selectvisco( visco ) 

gas=int(input( "Enter the % of free gas:  "))
selectgas( gas )



def offsh():
    global gl
    d=str(input("Does the well belong to offshore?(Y/N)"))
    while d not in ("Y","N","y","n"):
        clear_output(wait=True)
        d=str(input("Does the well belong to offshore?(Y/N)"))
        
    if(d=='Y' or d=='y'):
        srp[0]=-1
        pcp[0]=-1
    return

offsh()
a=0
cand=[None]*4
def bestcand():         #Fuction to check the best lift
    global gl,srp,esp,pcp,cand,a
    
    d=max((gl[0],esp[0],pcp[0],srp[0]))     
    if(d==gl[0]):
        cand[a]=gl
        a=a+1
    if(d==esp[0]):
        cand[a]=esp
        a=a+1
    if(d==pcp[0]):
        cand[a]=pcp
        a=a+1
    if(d==srp[0]):
        cand[a]=srp
        a=a+1   
    a=a-1
    return

bestcand()     #calling function to select best 

#calling function to select best 

while(a>-1):
    print("The artificial lift suitable is: ",cand[a][1])
    a = a-1
```
